```python
import random

def simulate_dice_rolls(num_rolls):
    sums_count = {i: 0 for i in range(2, 13)}  # словник для підрахунку кількості кожної суми

    for _ in range(num_rolls):
        dice1 = random.randint(1, 6)  # кидок першого кубика
        dice2 = random.randint(1, 6)  # кидок другого кубика
        total = dice1 + dice2
        sums_count[total] += 1  # оновлення кількості випадінь суми

    probabilities = {k: v / num_rolls * 100 for k, v in sums_count.items()}  # обчислення ймовірностей у відсотках
    return probabilities

# Виконати симуляцію
num_rolls = 1000000  # велика кількість кидків кубиків
simulated_probabilities = simulate_dice_rolls(num_rolls)

# Вивести результати симуляції
print("Сума\tІмовірність")
for total, probability in simulated_probabilities.items():
    print(f"{total}\t{probability:.2f}% ({simulated_probabilities[total]}/{num_rolls})")
```
